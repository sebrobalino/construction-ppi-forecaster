```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load data
data_path = Path("../data/WPUSI012011.csv")
df = pd.read_csv(data_path)
df.columns = ["date", "ppi"]  # rename columns for consistency
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date").reset_index(drop=True)

# Create input table
inputs = df.copy()
inputs["year_month"] = inputs["date"].dt.to_period("M").astype(str)
inputs["material"] = "Construction Materials"

# Simple baseline: predict next month = last observed
df["predicted_ppi"] = df["ppi"].shift(1)

```

```python
fig, ax = plt.subplots(figsize=(6, 2))
ax.axis("off")
tbl = ax.table(
    cellText=inputs[["year_month", "material"]].head(5).values,
    colLabels=["year_month", "material"],
    loc="center"
)
plt.savefig("../images/input_example.png", dpi=200, bbox_inches="tight")
plt.close()

```

```python
output_view = df[["date", "ppi", "predicted_ppi"]].dropna().head(5)
output_view["year_month"] = output_view["date"].dt.to_period("M").astype(str)

fig, ax = plt.subplots(figsize=(6, 2))
ax.axis("off")
tbl = ax.table(
    cellText=output_view[["year_month", "predicted_ppi"]].values,
    colLabels=["year_month", "predicted_ppi"],
    loc="center"
)
plt.savefig("../images/output_example.png", dpi=200, bbox_inches="tight")
plt.close()

```
